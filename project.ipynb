### 21- Find the subdataframe of the dfcode dataframe you created earlier on where the "Country Code" value is **not** in the "iso3" column of capitaldf. Hint: look up and use the .isin pandas Series method. Store this subdataframe in the variable dfnocode. Display dfnocode and explain in a markdown cell what you see and what this could be used for.

- ##### To Explain how this new dataframe could be use for, we do as follows:


```python
#countries in dfcode which their codes are not in iso3 of capitaldf 
codes = capitaldf['iso3']
filt = dfcode['Country Code'].isin(codes)
dfnocode = dfcode.loc[~filt]
dfnocode
```

##### there are 50 countries whith this specification. It means there are 50 country codes is in '***dfcode***' dataframe that are not in '***iso3***' column of '***capitaldf***' dataframe.  

```python
len(dfnocode)
```

#### On the other hand there are 80 countries in '***dfcode***' that are not in '***capitaldf***'
- we name the dataframe related to them '***dfnoname***' 

```python
# countris in dfcode and not in capitaldf
countries = capitaldf['name']
ctryDiff = dfcode['Country Name'].isin(countries)
dfnoname = dfcode.loc[~ctryDiff]
dfnoname
```

##### It seems there are 30 inconsistencies between names and codes of dfcode dataframe. Because 30 countries of '***dfcode***' dataframe have code in '***iso3***' column of '***capitaldf***' dataframe but have not name in the dataframe. 
- we find these 30 countries as bellow

```python
# A mask to filter countries in dfcode dataframe that has no name in capitaldf(countries in dfnoname) and have code in capitaldf ()
filt = (dfcode['Country Name'].isin(dfnoname['Country Name'])) & ~(dfcode['Country Code'].isin(dfnocode['Country Code']))
dfcode.loc[filt]
```

### we can use this to correct the country names in '***dfcode***' dataframe. countries that their names are writen differently from their equvalent names in '***capitaldf***' dataframe.
- ##### for example we can see this difference in country name for the code '***IRN***' in both dataframe as bellow

```python
capitaldf.loc[capitaldf['iso3'] == 'IRN'] 
```

```python
dfcode.loc[dfcode['Country Code'] == 'IRN'] 
```
